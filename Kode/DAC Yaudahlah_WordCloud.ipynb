```python
from wordcloud import WordCloud
wordcloud = WordCloud(width=1600, height=800, max_font_size=200, background_color='white')

import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_dana = pd.read_csv('/content/drive/Shareddrives/DAC UNPAD: Yaudahlah/data/processedV4/dana.csv')
df_link = pd.read_csv('/content/drive/Shareddrives/DAC UNPAD: Yaudahlah/data/processedV4/link.csv')
df_ovo  = pd.read_csv('/content/drive/Shareddrives/DAC UNPAD: Yaudahlah/data/processedV4/ovo.csv')
df_sakuku = pd.read_csv('/content/drive/Shareddrives/DAC UNPAD: Yaudahlah/data/processedV4/sakuku.csv')

# minimum = min([len(df_dana), len(df_link), len(df_ovo), len(df_sakuku)])

# df_dana   = df_dana.head(minimum)
# df_link   = df_link.head(minimum)
# df_ovo    = df_ovo.head(minimum)
# df_sakuku = df_sakuku.head(minimum)
```

# WordCloud Dana

```python
dana_word = {}
for i in range(len(df_dana)):
  temp_text = df_dana.loc[i]['processed']
  temp_split = temp_text.split(' ')
  temp_set = list(set(temp_split))
  for j in range(len(temp_set)):
    try:
      dana_word[temp_set[j]] += 1
    except:
      dana_word[temp_set[j]] = 1

dana_word_sorted = dict(sorted(dana_word.items(), key=lambda x: x[1], reverse=True))
dana_word_sorted
```

```python
wordcloud.generate_from_frequencies(dana_word_sorted)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
